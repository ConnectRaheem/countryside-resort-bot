```python
from dotenv import load_dotenv
load_dotenv()
```

Initialize the mode

```python

from langchain.chat_models import init_chat_model

llm = init_chat_model(
    model="llama-3.3-70b-versatile",  
    model_provider="groq",
    temperature=0.7  
)

print("Model loaded successfully!")
```

```python
response = llm.invoke("Hello! You are a helpful assistant for Countryside Resort in Gilgit. Introduce yourself briefly.")

print(response.content)
```
